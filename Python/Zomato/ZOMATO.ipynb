## Observation
1. when ratingis between 4.5 to 4.9 ----->Excellent
2. when ratingis between 4.0 to 3.9 ----->Very good
3. when ratingis between 3.5 to 3.9 ----->Good
4. when ratingis between 3.0 to 3.4 ----->average
5. when ratingis between 2.5 to 2.9 ----->average
6. when ratingis between 2.0 to 2.4 ----->poor

```python
rating.head(3)
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating_counting",data=rating)
```
